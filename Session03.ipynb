```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

## Qubit Entanglement Chaining (GHZ State [names of scientists])

```python
nqbits = 20
qc = QuantumCircuit(nqbits) #0 classical bits, as we will be using measure_all which is a macro; creatyes required cbits
```

### Setting 1 qbit to superposition

```python
qc.h(0)
```

Chain each qbit to its previous:

```python
for i in range(nqbits-1):
  qc.cx(i, i+1) #(control, target)
```

Collapse all qubits:

```python
qc.measure_all()
```

Run the simulation:

```python
sim = AerSimulator()
job = sim.run(qc, shots=100000)
counts = job.result().get_counts()
print(counts)
```

Notice only 000... and 111.. states are formed.
- The first qbit is in superposition (both 0,1 and neither 0,1 at the same time).
- We entangle the next qbit (qbit2) to this, qbit2 is now 100% same state as qbit1.
- Now qbit3 entangled with qbit2, which again gets same state.
- Thus on collapsing entire system, all qubits have same state as qbit1
