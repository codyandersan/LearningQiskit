## CNOT Gate

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
circuit = QuantumCircuit(2, 2)
```

Put first qubit in superposition (h gate)

```python
circuit.h(0)
```

#### Apply CNOT Gate:
(controlQbit, targetQbit)

```python
circuit.cx(0, 1) #first qubit's state controls second qubit now
```

```python
circuit.measure([0,1], [0,1]) #([qbit1, qbit2], [cbit1, cbit2])
```

#### Running the simulator:

```python
nshots = 100000
```

```python
simulator = AerSimulator()
job = simulator.run(circuit, shots=nshots)
counts = job.result().get_counts()
print(counts)
```

##### Result
Only 00, 11 states present, 01, 10 never exist as both qubits are entangled
qbit1 is random (superposition), but as soon as you measure it, qbit2 also gets the same measurement

#### Measuring both qubits without entanglement:

```python
def entangled_measures():
    circuit = QuantumCircuit(2, 2)
    circuit.h(0)
    circuit.cx(0, 1) #first qubit's state controls second qubit now

    circuit.measure(0, 0) #only 1st qubit measured

    simulator = AerSimulator()
    job = simulator.run(circuit, shots=nshots)
    counts1 = job.result().get_counts()
    print(counts1) #gave {'01': 49861, '00': 50139} for eg

    #Now collapsing second qubit too:
    circuit.measure(1, 1)
    job = simulator.run(circuit, shots=nshots)
    counts2 = job.result().get_counts()
    print(counts2) #gives {'11': 50006, '00': 49994} for eg (no 10, 01)

entangled_measured()
```

```python
def independent_measures():
    circuit = QuantumCircuit(2, 2)
    circuit.h(0)
    circuit.h(1) #new addition
    #circuit.cx(0, 1) not executed

    circuit.measure([0,1], [0,1]) #both qubits measured

    simulator = AerSimulator()
    job = simulator.run(circuit, shots=nshots)
    counts = job.result().get_counts()
    print(counts) #gave {'11': 24932, '01': 24864, '10': 24910, '00': 25294} for eg [ALL 4 POSSIBILITIES]


independent_measures()
```
