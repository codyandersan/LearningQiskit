```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import time
```

build a system/computer that has 1qbit, 1 classical bit (cbit to store result)

Circuit is only a blueprint, stores all instructions/gates to be applied. To actually run, we need simulator or a quantum computer

```python
circuit = QuantumCircuit(1, 1) #(qbit, cbit)
```

qbit is either 0/1

Apply Hadamard gate on first qbit (index zero)
H gate brings it to a superposition, both 0 and 1.

```python
simulator = AerSimulator()
job = simulator.run(circuit, shots=100000)
result = job.result()
counts = result.get_counts() #dict
```

Stress Testing:

```python
def stress_test(nqbits, nshots):
    circuit = QuantumCircuit(nqbits)
    #every qubit undergoes h gate
    for i in range(nqbits):
        circuit.h(i)
        
    circuit.measure_all() #collapse all qubits
    
    simulator = AerSimulator()
    start_time = time.time()
    job = simulator.run(circuit, shots=nshots)
    result = job.result()
    end_time = time.time()
    counts = result.get_counts() #dict
    print(counts)
    return (counts, end_time-start_time)
```

#### measure_all:
a macro; auto creates rqeuired cbits and assigns the collapsed value of qbit to them.
#### measure:
manually assign collapsed value to a cbit already created in circuit

```python
output = stress_test(10, 1024000)
```

```python
plot_histogram(output[0], bar_labels=False)
```

Played with the graph by changing shots and no of qubits
Y axis is the count of how many times a particular value appears shots/(2^n)
for n=5, ideally every value(0/1 permutation of 5 qubits) shall appear 100/32=3 times, graph gets more even on increasing shots

```python
output = stress_test(20, 100000)
plot_histogram(output[0])
```
