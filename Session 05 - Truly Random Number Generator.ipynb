#Truly Random Number Generator
- Classical methods of random number generation use algorithms which if all required states and conditions are knwo, can be reversed, thus predicting the generated number
- This code uses the concept of Quantum Superposition to guarantee that each number in the specifed range has an equal chance to be generated. The random number cannot be predicted beforehand by any means.
- For perfectly random numbers, use IBM's Free Quantum Computer API instead of AerSimulator.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import math
```

```python
'''n: no of bits of integer to generate (0 to 2^n)
'''
def generateCircuit(n):
  qc = QuantumCircuit(n) #add n qubits
  for i in range(n):
    qc.h(i) #put all into superposition
  qc.measure_all()
  return qc
```

```python
def runSimulation(qc):
  n = qc.num_qubits
  sim = AerSimulator()
  job = sim.run(qc, shots=1)
  counts = job.result().get_counts()
  return counts
```

```python
def generateInteger(n, start, stop):
  while True:
    qc = generateCircuit(n)
    count = runSimulation(qc)
    randomInteger = int(list(count.keys())[0], base=2) + start
    if randomInteger<=stop: #might generate vals out of range due to more bits than required
      return randomInteger
```

```python
def rangeToQbits(start, stop):
  range = stop-start+1
  qbits = math.ceil(math.log2(range))
  return qbits
```

```python
def main():
  start = int(input("Enter start value:\t"))
  stop = int(input("Enter stop value:\t"))
  n = rangeToQbits(start, stop)
  randomInt = generateInteger(n, start, stop)
  print("Your truly random integer is:", randomInt)

main()
```
