```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

build a system/computer that has 1qbit, 1 classical bit (cbit to store result)

```python
circuit = QuantumCircuit(1, 1) #(qbit, cbit)
```

qbit is either 0/1

Apply Hadamard gate on first qbit (index zero)
H gate brings it to a superposition, both 0 and 1. 

Applying x gate before H:

```python
circuit.x(0)
```

```python
circuit.h(0) #(index of qbit)
```

change to classical bit (observing/collapsing)

```python
circuit.measure(0, 0) #(qbitIndex, resultStoringCbitIndex)
```

Run on simulator
Count: how many times experiment is executed

```python
simulator = AerSimulator()
job = simulator.run(circuit, shots=1024)
result = job.result()
counts = result.get_counts() #dict
```

```python
print("Measurement Counts: ", counts)
```
